```python
# ライブラリのインポート（再起動後に実行）
from vllm import LLM, SamplingParams

llm = LLM(model="Qwen/Qwen2.5-32B-Instruct", gpu_memory_utilization=0.9)

# 推論テスト
sampling_params = SamplingParams(temperature=0.2, top_p=0.95, max_tokens=100)
outputs = llm.generate(["vLLMのセットアップが完了しました。一言挨拶してください。"], sampling_params)

for output in outputs:
    print(output.outputs[0].text)
```

```python
# 10個の異なる質問をリストにする
prompts = [
    "日本で一番高い山は？",
    "美味しいカレーの作り方を1行で。",
    "Pythonの魅力とは？",
    "空はなぜ青いの？",
    "猫と犬、どっちが飼いやすい？",
    "1+1は？",
    "AIと人間の違いは？",
    "おすすめの旅行先は？",
    "プログラミング上達のコツは？",
    "今日の晩御飯は何がいい？"
]

# 全員同時に生成開始！
outputs = llm.generate(prompts, sampling_params)

# 結果を表示
for output in outputs:
    prompt = output.prompt
    generated_text = output.outputs[0].text
    print(f"【問】: {prompt}\n【答】: {generated_text.strip()}\n{'-'*30}")
```
